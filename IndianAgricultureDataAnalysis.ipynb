###Which state has the highest production of rice and wheat in the most recent year?


```
df.groupby(['State_Name', 'Crop', 'Crop_Year']).sum()
```

```
state_crop = df.groupby(['State_Name', 'Crop', 'Crop_Year']).sum()
```

```
state_crop.reset_index(inplace=True)
```

```
state_crop.head()
```

```
state_crop.Crop_Year.max()
```

```
state_rice = state_crop[(state_crop.Crop == 'Rice') & (state_crop.Crop_Year == 2014)] #We avoid 2015 because only two states data are available for 2015
```

```
state_rice[state_rice.Production == state_rice.Production.max()]
```

```
state_wheat = state_crop[(state_crop.Crop == 'Wheat') & (state_crop.Crop_Year == 2014)]
```

```
state_wheat[state_wheat.Production == state_wheat.Production.max()]
```

In the same method above, we can find the states with max or min production or area of any particular crop in a particular year

```
state_crop[(state_crop.Crop == 'Rice') & (state_crop.Crop_Year == 2014)].head()
```

```
fig = plt.gcf()
fig.set_size_inches(10,7)
sns.barplot(x='State_Name', y='Production', data=state_rice, order=state_rice.sort_values('Production').State_Name)
plt.xticks(rotation=90)
plt.xlabel("State")
plt.title("Rice production statewise in 2014")
plt.show()
```

###Which are the 10 highest produced crops in India recently?Which were those in 2000?Are they same?

```
df.head(3)
```

```
df[df.Crop_Year.isin([2014,2000])].groupby(['Crop_Year','Crop'])['Production'].sum()
```

```
crop_prod = df[df.Crop_Year.isin([2014,2000])].groupby(['Crop_Year','Crop'])['Production'].sum()
crop_prod_10 = crop_prod.sort_values().tail(10)
crop_prod_10
```

```
crop_prod_10 = crop_prod_10.reset_index()
```

```
crop_prod_10
```

```
fig = plt.gcf()
fig.set_size_inches(10,7)
sns.barplot(x='Crop', y='Production', data=crop_prod_10, hue='Crop_Year')
plt.xticks(rotation=90)
plt.xlabel("Crops", fontdict={'size' : 14})
plt.ylabel("Production", fontdict={'size' : 14})
plt.title("Highest produced crops in 2014 vs 2000", fontdict={'size' : 16})
plt.grid()
plt.show()
```

The most produced 5 crops are the same in 2000 and in 2014.Those are Coconut,Sugarcane, Rice, Wheat and Potato. As coconut's weight is uncomparable to other crops we plot the remaining 4 crops in a seperate plot below for better understanding of data.

```
fig = plt.gcf()
fig.set_size_inches(10,7)
sns.barplot(x='Crop', y='Production', data=crop_prod_10[:-2], hue='Crop_Year')
plt.xticks(rotation=90)
plt.xlabel("Crops", fontdict={'size' : 14})
plt.ylabel("Production", fontdict={'size' : 14})
plt.title("Highest produced crops in 2014 vs 2000", fontdict={'size' : 16})
plt.grid()
plt.show()
```

From the above chart, it is clear that except for wheat , production of all other crops have increased from year 2000 to 2014.There is a considerable dip in the production of wheat.

From the above plot , Kerala seems to be very high in production with very low area of cultivation.This needs further investigation

```
df_.sort_values('Crop_Year', inplace=True)
```

```
df[(df.State_Name == "Kerala") & (df.Crop_Year == 2000)].sort_values('Production')
```

It can be seen that this disparity comes from coconut, which is the main crop produced in kerala. For a realistic view of data, we now take only the 4 major crops as below and redo the plot 

```
df_ = df[df.Crop.isin(['Rice', 'Wheat', 'Maize', 'Ragi'])].groupby(['State_Name', 'Crop_Year']).sum()
```

```
df_.head()
```

```
df_.reset_index(inplace=True)
```

```
df_.sort_values('Crop_Year', inplace=True)
```

```
px.scatter(df_, x="Area", y="Production", animation_frame="Crop_Year", 
           animation_group="State_Name", color="State_Name")
```

```
df_['Efficiency'] = df_['Production']/df_['Area']
```

```
px.scatter(df_, x="Area", y="Efficiency", size="Production", animation_frame="Crop_Year", 
           animation_group="State_Name", color="State_Name", range_y = [0.75, 5], range_x=[-1E6, 20E6])
```
